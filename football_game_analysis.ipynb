#One-Way-Anova

```python
import pandas as pd
import numpy as np
from scipy.special import beta
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
football[['gk_diving','gk_handling','gk_kicking','gk_reflexes','gk_speed','gk_positioning']]
```

```python
diving = football['gk_diving']
handling = football['gk_handling']
kicking = football['gk_kicking']
reflexes = football['gk_reflexes']
speed = football['gk_speed']
positioning = football['gk_positioning']
Goalkeeper = np.stack([diving, handling, kicking , reflexes , speed , positioning], axis=0)
```

```python
n, (a, n_i) = Goalkeeper .size, Goalkeeper .shape
print(f"There are {a} groups with {n_i} samples each for a total of {n} samples.")
grand_mean, group_means = Goalkeeper .mean(), Goalkeeper .mean(1)
```

```python
#collapse-hide
x = np.arange(n)
plt.plot(figsize=(600, 400))
plt.scatter(x, Goalkeeper.ravel())
plt.hlines(grand_mean, 0, n, alpha=0.5)
for i, m in enumerate(group_means):
    plt.hlines(m, x[n_i*i], x[n_i*(i+1)-1], colors='r', linestyles='--', alpha=0.7) 
plt.grid()
plt.ylim(0, 100)
plt.tick_params(labelsize=15)
plt.ylabel("GoalKeeper (actions)", fontsize=20)
plt.xlabel("Sample", fontsize=20);
```

### Calculating F-ratio.

```python
def f_distribution(x, d_b, d_w):
    "F-distribution given `d_b` between DF and `d_w` within DF."
    return (d_b/d_w)**(d_b/2)*x**(d_b/2-1)*(1+x*d_b/d_w)**(-(d_b+d_w)/2)/beta(d_b/2, d_w/2)
```

```python
ss_within = np.sum((Goalkeeper - group_means[:, None])**2)
ss_between = np.sum(n_i*(group_means - grand_mean)**2)
df_within, df_between = n - a, a - 1
ms_within, ms_between = ss_within/df_within, ss_between/df_between
f_ratio = ms_between/ms_within
print(f"F-ratio of {f_ratio:.4f}.")
```

```python
fs = np.linspace(f_ratio, 1000, 10000)
prob = f_distribution(fs, df_between, df_within)
p_value = np.trapz(prob, x=fs)
print(f"p-value {p_value:.4f}")
```

The obtained p-value is significantly higher than the typical  0.05  threshold, so we cannot accept the hypothesis that there are significant differences between actions.

#Two Way Anova

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Suppress Warnings (not in the blog post)
import warnings
warnings.filterwarnings("ignore")

from scipy import stats
```

```python
football[['weak_foot','preferred_foot','skill_moves']]
```

```python
#len supp dose
#x='dose', y='len', hue='supp',
ax = sns.pointplot(x='skill_moves', y='weak_foot', hue='preferred_foot', data=football)
```

```python
N = len(football.weak_foot)
df_a = len(football.preferred_foot.unique()) - 1
df_b = len(football.skill_moves.unique()) - 1
df_axb = df_a*df_b 
df_w = N - (len(football.preferred_foot.unique())*len(football.skill_moves.unique()))
```

```python
grand_mean = football['weak_foot'].mean()
```

```python
# Sum of Squares A
ssq_a = sum([(football[football.preferred_foot == l].weak_foot.mean() - grand_mean)**2 for l in football.preferred_foot])
# Sum of Squares B
ssq_b = sum([(football[football.skill_moves ==l].weak_foot.mean() - grand_mean)**2 for l in football.skill_moves])
# Sum of Squares T
ssq_t = sum((football.weak_foot - grand_mean)**2)
```

```python
left = football[football.preferred_foot == 'LEFT']
right = football[football.preferred_foot == 'RIGHT']
left_skill_moves_means = [left[left.skill_moves == d].weak_foot.mean() for d in left.preferred_foot]
right_skill_moves_means = [right[right.preferred_foot == d].weak_foot.mean() for d in right.preferred_foot]
ssq_w = sum((right.weak_foot - right_skill_moves_means)**2) +sum((left.weak_foot - left_skill_moves_means)**2)
# Interaction
ssq_axb = ssq_t - ssq_a - ssq_b - ssq_w
```

```python
ms_a = ssq_a/df_a
ms_b = ssq_b/df_b
ms_axb = ssq_axb/df_axb
ms_w = ssq_w/df_w
```

```python
f_a = ms_a/ms_w
f_b = ms_b/ms_w
f_axb = ms_axb/ms_w
```

```python
p_a = stats.f.sf(f_a, df_a, df_w)
p_b = stats.f.sf(f_b, df_b, df_w)
p_axb = stats.f.sf(f_axb, df_axb, df_w)
```

```python
results = {'sum_sq':[ssq_a, ssq_b, ssq_axb, ssq_w],
           'df':[df_a, df_b, df_axb, df_w],
           'F':[f_a, f_b, f_axb, 'NaN'],
            'PR(>F)':[p_a, p_b, p_axb, 'NaN']}

columns=['sum_sq', 'df', 'F', 'PR(>F)']

aov_table1 = pd.DataFrame(results, columns=columns,
                          index=['preferred_foot', 'skill_moves', 
                          'preferred_foot:skill_moves', 'Residual'])
```

```python
def eta_squared(aov):
    aov['eta_sq'] = 'NaN'
    aov['eta_sq'] = aov[:-1]['sum_sq']/sum(aov['sum_sq'])
    return aov

def omega_squared(aov):
    mse = aov['sum_sq'][-1]/aov['df'][-1]
    aov['omega_sq'] = 'NaN'
    aov['omega_sq'] = (aov[:-1]['sum_sq']-(aov[:-1]['df']*mse))/(sum(aov['sum_sq'])+mse)
    return aov


eta_squared(aov_table1)
omega_squared(aov_table1)
print(aov_table1)
```
